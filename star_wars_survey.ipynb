# Star Wars Survey

## Introduction

In this project will be cleaning and exploring the surveys made by [FiveThirtyEight](https://fivethirtyeight.com/). The survey is about the question: **does the rest of America realize that "The Empire Strikes Back" is clearly the best of all Star Wars movies ever made?**. They surveyed *Star Wars* fans using the online tool SurveyMonkey. They received 835 total responses which you can download from [their Github repository](https://github.com/fivethirtyeight/data/tree/master/star-wars-survey).  


The data has several columns, including:

  * RespondentID - An anonymized ID for the respondent (person taking the survey)
  * Gender - The respondent's gender
  * Age - The respondent's age
  * Household Income - The respondent's income
  * Education - The respondent's education level
  *  Location (Census Region) - The respondent's location
  * Have you seen any of the 6 films in the Star Wars franchise? - Has a Yes or No response
  * Do you consider yourself to be a fan of the Star Wars film franchise? - Has a Yes or No response

```python
import pandas as pd

star_wars = pd.read_csv('star_wars.csv', encoding='ISO-8859-1')
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars.shape
```

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.shape
```

## Cleaning and Mapping Yes/No Columns

```python
yes_no = {
    "Yes": True,
    "No": False
}

question_cols =['Have you seen any of the 6 films in the Star Wars franchise?','Do you consider yourself to be a fan of the Star Wars film franchise?']
for cols in question_cols:
    star_wars[cols] = star_wars[cols].map(yes_no)

star_wars.head(10)
```

## Cleaning and Mapping Checkbox Columns

```python
import numpy as np

movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
```

```python
star_wars = star_wars.rename(columns={
        "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
        "Unnamed: 4": "seen_2",
        "Unnamed: 5": "seen_3",
        "Unnamed: 6": "seen_4",
        "Unnamed: 7": "seen_5",
        "Unnamed: 8": "seen_6"
        })
star_wars.head(10)
```

## Cleaning the Ranking Columns

```python
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    'Unnamed: 10': 'ranking_2',
    'Unnamed: 11': 'ranking_3',
    'Unnamed: 12': 'ranking_4',
    'Unnamed: 13': 'ranking_5',
    'Unnamed: 14': 'ranking_6'
})

print(star_wars.columns)
print(star_wars.head(10))
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

## Finding the Highest-Ranked Movie

```python
star_wars_ranking_mean = star_wars[star_wars.columns[9:15]].mean()

%matplotlib inline
import matplotlib.pyplot as plt

star_wars_ranking_mean.plot.bar(rot=30)
```

## Rankings

So far, we've cleaned up the data, renamed several columns, and computed the average ranking of each movie. As suspected, it looks like the ranking_5 is rated as the best movie among the survey.

## Finding the Most Viewed Movie

```python
seen_sum = star_wars[star_wars.columns[3:9]].sum()
seen_sum.plot.bar(rot=30)
```

## View counts

It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular. Also based on the graph above we can tell that the reason behind the fifth movie was rated as the highest rank it garnered around over 750 viewers out of 835 who take the survey.  

## Exploring the Data by Binary Segments

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
males_mean = males[males.columns[9:15]].mean()
males_mean.plot.bar(rot=30)
```

```python
females_mean = females[females.columns[9:15]].mean()
females_mean.plot.bar(rot=30)
```

```python
males_seen = males[males.columns[3:9]].mean()
males_seen.plot.bar(rot=30)
```

```python
females_seen = females[females.columns[3:9]].mean()
females_seen.plot.bar(rot=30)
```

Based on the observations above, there are more male views for Star Wars Movies 1-3 than females.
